# 모듈불러오기

```python
import urllib.request as req
import urllib.parse

target_site = 'https://ko.wikipedia.org/wiki/%EB%B8%8C%EB%9D%BC%EC%A7%88'
res = req.urlopen(target_site)
```

```python
import bs4
from bs4 import BeautifulSoup

bs4.__version__
```

```python
soup = BeautifulSoup(res, 'html5lib')
```

# 데이터 추출

## 주 이름 추출하기

```python
# 1-1. li 코드 추출하기
li_code = soup.select('div.mw-content-ltr.mw-parser-output > table > tbody > tr > td > ol > li')
li_code
```

```python
# 1-2. state_name 추출하기
state_name = []
for i in range(0, len(li_code)):
    state_name.append(li_code[i].get_text())

state_name, len(state_name)
```

## 우편약자 추출하기

- 각 페이지로 이동하는 코드 생성
- 해당 페이지에서 우편약자 추출

```python
# 각 페이지로 이동하는 코드 생성하기
base_url = 'https://ko.wikipedia.org/wiki/'
urls = [base_url + urllib.parse.quote(name) for name in state_name]
```

```python
# 해당 페이지에서 우편약자 추출하기
post_codes = []

for url in urls:
    try:
        page_response = req.urlopen(url)
        page_soup = BeautifulSoup(page_response, 'html5lib')

        post_code_selector = '#mw-content-text > div.mw-content-ltr.mw-parser-output > table.infobox.bordered > tbody > tr:nth-child(12) > td'
        post_code_element = page_soup.select_one(post_code_selector)
        if post_code_element:
            post_code = post_code_element.get_text(strip=True)
            post_codes.append(post_code)
        else:
            post_codes.append(url, 'post_code를 찾을수 없습니다')
    except Exception as e:
        post_codes.append((url, f"오류발생 : {e}"))

for post_code in post_codes:
    print(post_code)
```

## 브라질식 표기 추출

```python
selector_1 = '#mw-content-text > div.mw-content-ltr.mw-parser-output > table:nth-child(88)'
selector_2 = '#mw-content-text > div.mw-content-ltr.mw-parser-output > table:nth-child(89)'
element_1 = soup.select_one(selector_1)
element_2 = soup.select_one(selector_2)

# 선택한 요소의 HTML 콘텐츠를 문자열로 추출
html_content_1 = str(element_1)
html_content_2 = str(element_2)

# HTML 콘텐츠 출력
print(html_content_1)   # 마투그로수두술은 영문표기가 없음 -> 임의로 추가예정 (Mato Grosso do Sul)
print(html_content_2)
```

```python
import re

# 대괄호 안의 텍스트를 추출하는 함수
def extract_text_in_brackets(text):
    matches = re.findall(r'\[([^\]]+)\]', text)
    return matches

name_text_1 = extract_text_in_brackets(html_content_1)
name_text_2 = extract_text_in_brackets(html_content_2)
```

```python
# name_text_1의 마지막에 "Mato Grosso do Sul" 추가
name_text_1.append("Mato Grosso do Sul")
print(name_text_1)
```

```python
# name_text_1 과 name_text_2 합치기
name_text = name_text_1 + name_text_2
print(name_text, len(name_text))    # 27
```

# 데이터 변환 후 병합

```python
import numpy as np
import pandas as pd
```

## state_name 주이름

```python
state_name = pd.DataFrame(state_name)
state_name.columns = ['주이름']
state_name['주이름'] = state_name['주이름'].str.replace(' ', '')
state_name
```

## post_codes 우편약자

```python
post_codes = pd.DataFrame(post_codes)
post_codes.columns = ['우편약자']
post_codes.loc[2, '우편약자'] = 'AM'
post_codes.loc[3, '우편약자'] = 'PA'
post_codes.loc[15, '우편약자'] = 'BA'   # 페이지에 잘못된 정보로 나와있어서 수정
post_codes.loc[17, '우편약자'] = 'GO'
post_codes.loc[18, '우편약자'] = 'DF'
post_codes.loc[22, '우편약자'] = 'RJ'
post_codes.loc[23, '우편약자'] = 'SP'
post_codes.loc[24, '우편약자'] = 'PR'
post_codes.loc[25, '우편약자'] = 'SC'
post_codes
```

## name_text 브라질식 표기

```python
name_text = pd.DataFrame(name_text)
name_text.columns = ['브라질식 표기']
name_text
```

## eng_text 영어식 표기

- 변환 내용
    - á, ã -> a
    - ô -> o
    - í -> i

```python
eng_text = name_text.copy()
# 문자열 대체하기
eng_text.columns = ['영어식 표기']
eng_text['영어식 표기'] = eng_text['영어식 표기'].str.replace('á', 'a')
eng_text['영어식 표기'] = eng_text['영어식 표기'].str.replace('ã', 'a')
eng_text['영어식 표기'] = eng_text['영어식 표기'].str.replace('ô', 'o')
eng_text['영어식 표기'] = eng_text['영어식 표기'].str.replace('í', 'i')
eng_text
```

## 병합

```python
# 파일병합하기
brazil_state = pd.concat([state_name, post_codes, name_text, eng_text], axis=1)
brazil_state
```

# 저장하기

```python
brazil_state.to_csv('brazil_state.csv', index=False)
```
