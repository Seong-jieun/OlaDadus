```python
import folium
import pandas as pd
from folium.plugins import HeatMap
```

```python
loc = pd.read_csv('/content/drive/MyDrive/물류 창고 추정 위치.csv')
loc
```

```python
cus = pd.read_csv('/content/customers_locations.csv')
cus.head()
```

```python
sell = pd.read_csv('/content/sellers_locations.csv')
sell
```

```python
lat = []
lng = []

for i in cus['고객 위도']:
  lat.append(i)
for i in cus['고객 경도']:
  lng.append(i)
```

```python
def locate():

  m = folium.Map(location = [-22.650880052023705, -47.5373834739126], zoom_start = 5)

  heatMap = HeatMap(zip(lat, lng),
                    radius = 10,
                    color = 'red')
  m.add_child(heatMap)

  for _, row in loc.iterrows():
    folium.Marker(location = [row['Latitude'], row['Longitude']], icon = folium.Icon(color = 'red', icon = 'home')).add_to(m)
  return m
```

```python
locate()
```
